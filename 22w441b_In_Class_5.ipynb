# Choose 4 features you think are important. Case_status is your target variable


```python
df_select = df[["case_status", "class_of_admission", "country_of_citizenship", "application_type", "wage_offered_from_9089"]]
df_select.head()
```

```python
df_select.info()
```

# Clean your data for a decision tree


```python
df_select.dropna(inplace=True)
df_select.reset_index(inplace=True)
df_select
```

```python
# Prepare to use LabelEncoder to label every categorical variable
from sklearn.preprocessing import LabelEncoder
```

```python
column_names = list(df_select.columns)
column_names.remove('index')
column_names.remove('wage_offered_from_9089')

le_dict = {}
for column in column_names:
  # Initialize the model
  le = LabelEncoder()

  # Label the categorical variables
  df_select[f"{column}_labeled"] = le.fit_transform(df_select[column])

  # Store the label encoder in a dictionary
  le_dict[column] = le

df_select.head()
```

```python
# Find X and y for model training
X = df_select[['class_of_admission_labeled','country_of_citizenship_labeled','application_type_labeled','wage_offered_from_9089']]
y = df_select[['case_status_labeled']] 
```

# Fit and plot a decision tree of depth 3


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
# Fit the decision tree model
clf = DecisionTreeClassifier(max_depth=3)
clf.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(25,10))
tree.plot_tree(clf, filled=True)
plt.show()
```

# Write your interpretation of the largest (by sample size) leaf node

The sample size of the largest leaf node indicates the confidence of the decision tree in its categorizing for that particular group of samples.

In this case it is the 8th among the results. It contains 65257 samples with a gini impurity of 0.591, which is so high that it indicates our variable selection failed.

# Using a for loop, make your own train-test split and determine the best max_depth for out-of sample accuracy

```python
from sklearn.model_selection import train_test_split
```

```python
outputs = []
max_depths = [i for i in range(1,10)]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
for md in max_depths:
  clf = DecisionTreeClassifier(max_depth=md)

  # Train the model
  clf.fit(X_train, y_train)

  # Test the model with out of sample data
  score = clf.score(X_test, y_test)
  outputs.append(score)
print(outputs)
```

According to the results, the highest accuracy is apporximately 47.34%, the max_depth is 8.
